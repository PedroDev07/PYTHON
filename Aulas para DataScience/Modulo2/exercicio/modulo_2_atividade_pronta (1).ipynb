```python
#Passando o filme que estava em 2° lugar no ranking para 1° usando o insert para inserir no elemento 0 e o pop para remover a duplicação no elemento 2

filmes = ['Um sonho de liberdade','O poderoso chefão','O poderoso chefão II','Batman: O cavaleiro das trevas','12 homens e uma sentença','A lista de schindler','O senhor dos anéis: O retorno do rei','Pulp fiction: tempo de violência','Tres homens em conflito','O senhor dos anéis: a sociedade do anel']

filmes.insert(0,'O poderoso chefão')
filmes.pop(2)
print ('----------------------------------------------------->')
print (filmes)
print ("Temos um total de",len(filmes),'filmes na lista')
print ('----------------------------------------------------->')


```
